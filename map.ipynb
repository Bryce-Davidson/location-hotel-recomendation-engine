```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import branca.colormap as cm
import folium
```

```python
# load the data
df_locations = pd.read_csv('./locations.csv')
df_hotels = pd.read_csv('./hotels.csv')
```

```python
def find_bounding_box_center(df: pd.DataFrame):
    lat_min, lat_max = df['lat'].min(), df['lat'].max()
    lng_min, lng_max = df['lng'].min(), df['lng'].max()
    center_lat = (lat_min + lat_max) / 2
    center_lng = (lng_min + lng_max) / 2
    return [center_lat, center_lng]
```

```python
# create a map

# disable controls
m = folium.Map(location=find_bounding_box_center(df), zoom_start=12, zoom_control=False)

# add markers to the map
for i, row in df_locations.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=3,
        popup=row['name'],
        stroke=True,
        fill=True,
        fill_color='blue',
        color='white',
        weight=1,
        fill_opacity=0.7
    ).add_to(m)

for i, row in df_hotels.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=3,
        popup=row['name'],
        stroke=True,
        fill=True,
        fill_color='red',
        color='white',
        weight=1,
        fill_opacity=0.7
    ).add_to(m)

# disable zoom
m.scroll_wheel_zoom = False

m
```
